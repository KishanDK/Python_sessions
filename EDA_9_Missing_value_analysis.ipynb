**Missing Value Analysis**

```python
import numpy as np
import pandas as pd
```

```python
dict1 = {'Name': ['Ram','Sita','Laxman',np.nan],
        'Age': [30,31,np.nan,33],
'city': [np.nan,'Hyd','Pune','Chennai']}
pd.DataFrame(dict1)
```

```python
df = pd.DataFrame(dict1)
df
```

$Method-1$:
**Fill with a random number to all null values**

```python
df.isnull()   # where ever null values are there it shows as True
```

```python
df.isnull().sum()
```

```python
df.fillna(40)      
```

```python
df
```

```python
df.fillna(40,inplace = True)  # inplace = True    it means it is updating the values to the dataframe
```

```python
df
```

```python
df.dtypes
```

$Method-2$:


**Fill the Random values by using column wise**

```python
########################   Read the data again     ###############################
dict1 = {'Name': ['Ram','Sita','Laxman',np.nan],
        'Age': [30,31,np.nan,33],
'city': [np.nan,'Hyd','Pune','Chennai']}
pd.DataFrame(dict1)
```

```python
#  when doing 'df[col].method(value, inplace=True)', try using 'df.method({col: value}, inplace=True)'   (error message)
dict1 = {'Name': ['Ram','Sita','Laxman',np.nan],
        'Age': [30,31,np.nan,33],
'city': [np.nan,'Hyd','Pune','Chennai']}
pd.DataFrame(dict1)
```

```python
df = pd.DataFrame(dict1)
df
```

```python
df['Name'] = df['Name'].fillna('Raheem')
df['Age'] = df['Age'].fillna(32)
df['city'] = df['city'].fillna('Blr')
df
```

$Method-3$:
- Pad
- bfill
- backfill
- ffill

```python
########################   Read the data again     ###############################
dict1 = {'Name': ['Ram','Sita','Laxman',np.nan],
        'Age': [30,31,np.nan,33],
'city': [np.nan,'Hyd','Pune','Chennai']}
df = pd.DataFrame(dict1)
```

```python
df
```

**Pad**

```python
print('============   Original   =============')
print(df)
print('================Pad===============')
print(df.fillna(method = 'pad'))
```

**bfill**

```python
print('============   Original   =============')
print(df)
print('================bfill===============')
print(df.fillna(method = 'bfill'))
```

**backfill**

```python
print('============   Original   =============')
print(df)
print('================backfill===============')
print(df.fillna(method = 'backfill'))
```

```python
print('============   Original   =============')
print(df)
print('================ffill===============')
print(df.fillna(method = 'ffill'))
```

$Method-4$:

- Mean and median is used for numerical data
