- Converting skewed distribution to normal distribution
- All maths developed by assumptions as data fallows normal distribution
    - log transformation
    - reciprocal transformation
    - sqrt transformation
    - exponential transformation
    - box-cox transformation
    - yeo-jhanosn transformation

$Step-1$
- import the packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

$Step-2$
- Read the data

```python
data = np.random.exponential(size = 1000)       # we are taking 1000 random sample values which are exponential
data
```

```python
plt.hist(data,bins = 40,label = 'Exponential')
plt.legend()
plt.show()
```

$Step-3$
- ***Log Transformation***
    - np.log represents natural logarithm
    - natural logarithm means base e
    - exponential will multiply with log base e
    - Natural logarithms will works positive data
    - log transformation will remove the skew
    - it will not convert into normal distribution

```python
log_data = np.log(data)
plt.hist(log_data,bins = 40, label = 'Log Transformation')
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1).hist(data,bins = 40)
plt.subplot(1,2,2).hist(log_data,bins = 40)
plt.show()
```

```python
log_log_data = np.log(log_data)
plt.hist(log_log_data,bins = 40, label = 'Log_Log Transformation')
plt.legend()
plt.show()
```

$step-4$
- reciprocal transformation
- if data has zero, it will fail

```python
reci_data = 1/data
plt.hist(reci_data,bins = 40, label = 'Reciprocal Transformation')
plt.legend()
plt.show()
```

$Step-5$
**Sqrt Transformation**

```python
sqrt_data = np.sqrt(data)
plt.hist(sqrt_data,bins = 40, label = 'Sqrt Transformation')
plt.legend()
plt.show()
```

$Power Transformer$

- it is in skleran.preprocessing
- method argument
- box-cox
- ye-jhonson

```python

from sklearn.preprocessing import PowerTransformer
data = data.reshape(-1, 1) 
pt = PowerTransformer(method='yeo-johnson')
data_transformed = pt.fit_transform(data)
data_transformed
```

```python

plt.hist(data_transformed,bins = 40, label = 'Power Transformation')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
data = data.reshape(-1, 1) 
pt = PowerTransformer(method='box-cox')
transformed_data = pt.fit_transform(data)
transformed_data
```

```python
plt.hist(transformed_data,bins = 40, label = 'Power Transformation(box-cox)')
plt.legend()
plt.show()
```
