## Step-1: Integer

```python
x = 15
print("Value of x:", x)
print("Type of x:", type(x))
print("x * 5 =", x * 5)
```

## Step-2: Float

```python
y = 12.75
print("Value of y:", y)
print("Type of y:", type(y))
print("y / 2 =", y / 2)
```

## Step-3: String

```python
name = "Python Programming"
print("String:", name)
print("First character:", name[0])
print("Last character:", name[-1])
print("Length of string:", len(name))
print("Substring (first 6 chars):", name[:6])
```

## Step-4: Boolean

```python
a = True
b = False
print("a:", a, "Type:", type(a))
print("b:", b, "Type:", type(b))
print("a and b:", a and b)
print("a or b:", a or b)
print("not a:", not a)
```

## Step-5: Complex Numbers

```python
z = 4 + 5j
print("z:", z)
print("Type of z:", type(z))
print("Real part:", z.real)
print("Imaginary part:", z.imag)
print("Conjugate:", z.conjugate())
```
