```python
# Generalized expression
labels = visa_df['continent'].unique()
Certified_counts = []
for i in labels:
    cond1 = visa_df['continent'] == i
    cond2 = visa_df['case_status'] == 'Certified'
    con = cond1&cond2
    Certified_counts.append(len(visa_df[con]))
labels, Certified_counts
```

```python
pd.DataFrame(zip(labels,Certified_counts),columns = ['continent','case_status'])
```

```python
pd.DataFrame(zip(labels,Certified_counts),columns = ['continent','case_status']).set_index('continent')
```

```python
# Generalized expression
labels = visa_df['continent'].unique()
Certified_counts = []
Denied_counts = []
for i in labels:
    cond1 = visa_df['continent'] == i
    cond2 = visa_df['case_status'] == 'Certified'
    cond3 = visa_df['case_status'] == 'Denied'
    Certified_counts.append(len(visa_df[cond1&cond2]))
    Denied_counts.append(len(visa_df[cond1&cond3]))
labels, Certified_counts, Denied_counts
```

```python
pd.DataFrame(zip(labels,Certified_counts,Denied_counts),columns=['continent','Certified','Denied'])
```
