**Onehot Encoder**
- One hot Encoder means at a time one will be on (1/True) and another one will be off(0/False)
- Suppose case_status has two unique labels
- "Certified" and "Denied"
- In One hot Encoder, it wil create two new column
- like below
case_status|Certified|Denied|
|----------|---------|------|
Certified|1|0|
Denied|0|1|

**Advantage:**
- One hot Encoder new columns are Orthogonal each other
- Orthoganality means 90 degrees phase shift
- 90 degrees phase shift means perpendicular each other
- perpendicular means independent to each other
- Independent means no relation
- It is very important property that is to avoid relation between variables
- **Draw back:**
- Assume that you have 100 uniwqe labels for a column
- So it will create 100 new columns
- To process 100 columns we require more memory, more time, more Hardware
- This is called **curse of Dimensionality**
